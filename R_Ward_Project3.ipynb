### Step 1: Read in the File 
To solve this I will first import Pandas and Numpy. Then I will read in the file using urllib.request and pass the link to the github file where the .txt file is stored. Next I will create an empty list and append each line of the text file to it.   

```python
#first import pandas
import pandas as pd

#also import numpy just in case
import numpy as np

#give the GitHub file a variable so that we can read in the file
filename = "https://raw.githubusercontent.com/rachelward617/DAV-5400/master/tournamentinfo.txt"

#now read in the file using urllib.request function
import urllib.request
data = urllib.request.urlopen(filename)

#create an empty list to which we will append each line of the file we just read in
tournament= []

#use a for loop to append each line of data as a string to the empty list created above
for line in data:
    tournament.append(str(line))

#print results
print (tournament)
```

### Step Two: Clean The Imported Data
Next I will use regex to remove the lines that are full of dashes which will make our data easier to read and analyze. Then I will convert the data into a DataFrame which will facilitate steps 3 through 5.

```python
#import regex library
import re

#define a regex that has at least two dashes in it
new="[-]{2}"

#use a for loop to iterate through the list and remove any items that match the regex above
for item in tournament:
    #if search does not give back 'None' then it is a match, so I will use boolean logic to tell the loop what to remove
    if re.search(new,item) != None:
        #remove if a match (i.e. search did not return 'None')
        tournament.remove(item)

#show results
print(tournament)
```

Now that I have removed the lines that only contained dashes, I will make this list into a Panda Series which will further facilitate the analysis.

```python
#now make this into a Pandas Series so it is easier to analyze
tourneyframe = pd.Series(tournament)

#return results
tourneyframe
```

### Step Three: Extract Desired Data: Names

Now that we have a cleaner data frame to work with, I will extract the player names first using a regex to capture first and last names, and put these in separate columns of a dataframe using the extract function from Pandas. Since this will also return the names of states, I will then create a concatenated full name column which concatenates the first name and last name. I will count the length of the strings in this column and remove the rows where the length is 3 because those rows contain state information. 

I will then store only the full names in a data frame and reindex this data frame so that when I extract more data further on, the new data can be easily concatenated to this "clean" data frame.

```python
#define a regex to capture the names of the players
pattern = r'([A-Z]+)([ A-Z]{2,})'

#define a new dataframe that will formed by extracting the names of the players using the regex above
names = tourneyframe.str.extract(pattern)

#rename the columns for ease of use
names = names.rename(columns={0: "First Name", 1: "Last Name"})

#return results
names.head()
```

```python
#now create a new column with the full player name so that we can analyze the length of that and remove the states
#to do this I will concatenate the First Name and Last Name columns
names["Player_Name"] = names["First Name"] + names["Last Name"]

#return results
names.head()
```

```python
#now create a column that counts the length of the player name column
names['Name_Count'] = names['Player_Name'].str.len()

#show the results
names.head(15)
```

```python
#remove the rows that have state information so that we can get a dataframe with just names
#from the above results we can see that the rows with State information have a name_count value of 3
#create a new frame which gives only rows where the name_count is not 3
namesclean = names[names.Name_Count != 3.0]

#show results
namesclean.head()
```

```python
#create a dataframe with ONLY full names - no count or first name, last name
fullnameonly = namesclean[['Player_Name']].copy()

#show results
fullnameonly.head()
```

```python
# Delete first two rows because one is a NaN and the other is a header name which we won't be using right now
fullnameonlyclean = fullnameonly.drop([0 , 1])

#return results
fullnameonlyclean.head()
```

### Step 3: Extract Desired Data: Total Points

Now that we have player name and state I will extract the player's total points.

To do so, I will create a regex for floating numbers and extract any items that match this regex because all of the total points contain decimals. Since this will likely return some numbers that do not contain decimal points, I will create a separate column which looks for a decimal point. Then I will extract only the rows which contain a decimal point in this column.

I will create a clean dataframe that contains only the total points, and reset the index of it as I did with the player name and state data frames above. Then I will concatenate my results dataframe and the new total points data frame.

```python
#create a regex that looks for float numbers
floating = r'([0-9].[0-9])'

#extract items that match this regex to a new dataframe
totalpoints = tourneyframe.str.extract(floating)

#show results
totalpoints.head()
```

```python
#define a regex to determine if a number has a decimal point in it
decimalpoint = r'(\.)'

#create a new column which will have a . in it ONLY when the float has a decimal
totalpoints['has_decimal'] = totalpoints[0].str.extract(decimalpoint)

#show results
totalpoints.head()
```

```python
#create a new dataframe which contains only rows where the has_decimal column has a decimal in it
#this will give us only the total points data which is what we want
decimalsonly = totalpoints[totalpoints.has_decimal == '.']

#show results
decimalsonly.head()
```

```python
#create a dataframe with ONLY total points i.e. does not have a has_decimal column
totalpointsonly = decimalsonly[[0]].copy()

#next we will reset the index so that later on we can make a combined data frame
#to do so we will use the reset_index function and drop the old index numbers
#this prevents the old index numbers from being made into a column
totalpointsonly.reset_index(drop=True, inplace = True)

#rename column to Total_Points
totalpointsonly = totalpointsonly.rename(columns={0: "Total_Points"})

#show results
totalpointsonly.head()
```

### Step 3: Extract Desired Data: Player Pre-Rating

Now that we have player name, state, and total points, I will extract the player's pre-rating.

To do so, I will create a regex to extract anything after the : symbol and before the - symbol. Then I will split the string to create a new column for anything following the P. I do this because after reading about Elo's chess rating system [at this link](https://medium.com/cantors-paradise/the-mathematics-of-elo-ratings-b6bfc9ca1dba), the chess rating is derived from a mathematical calculation and therefore cannot contain any letters.  

I will create a new dataframe that contains only the numbers up to the letter P, and then add this to our results dataframe which already contains player name, state, and total points.

```python
#create a regex to capture anything following the : up to the - 
pattern = r':(.*)\-'

#create a new dataframe containing this extracted information
prerate = tourneyframe.str.extract(pattern)

#show results
prerate.head()
```

```python
#now I will use the str.split() function to split the data into two columns
#containing digits to the left of P and right of P respectively
# dropping null value columns to avoid errors 
prerate.dropna(inplace = True) 

# new data frame with split value columns on the value P
newprerate = prerate[0].str.split("P", n = 1, expand = True) 
  
# make a separate column containing numbers before the letter P from new data frame 
prerate["Pre_Rating"]= newprerate[0] 
  
# make a separate column containing numbers from after the letter P from new data frame 
prerate["Garbage"]= newprerate[1] 
    
# display results
prerate.head(10)
```

```python
#create a dataframe with ONLY pre-rating 
prerateclean = prerate[['Pre_Rating']].copy()

#next we will reset the index so that later on we can make a combined data frame
#to do so we will use the reset_index function and drop the old index numbers
#this prevents the old index numbers from being made into a column
prerateclean.reset_index(drop=True, inplace = True)

#show results
prerateclean.head()
```

```python
#now we will add the pre-rating to our combined player name, state, and total points data frame
#make a list of the frames to concatenate
includeprerate = [result2,prerateclean]

#concatenate the series into a data frame with player name, state, total points, and pre rating in separate columns
result3 = pd.concat(includeprerate, axis =1)

#show the results
result3.head()
```

### Step 5: Export to a CSV

Now that all of the desired data is in a dataframe, I will export the dataframe to a .csv file. 

Since I want my code to be able to be replicated, I researched online for ways to export to the user's computer rather than the typical to_csv() options which would send the file to my own computer's hard drive.

I found a great website that had code built for this exact situation called [datatofish.com](https://datatofish.com/export-dataframe-to-csv/). I modified the code there to call my data frame.

If you hit run in the code cell below, an icon will pop up on your screen asking you where you would like to save the file and what you would like to call it.

```python
#import the tkinter module 
import tkinter as tk

#import the filedialog package
from tkinter import filedialog

#rename our dataframe with results to df for ease of use
df = finalresult

#the code from here on out I left untouched
#it sets up what the popup window that prompts you to save your csv will look like
#including the button that you click on to tell it where to save the csv file
root= tk.Tk()

canvas1 = tk.Canvas(root, width = 300, height = 300, bg = 'lightsteelblue2', relief = 'raised')
canvas1.pack()

def exportCSV ():
    global df
    
    export_file_path = filedialog.asksaveasfilename(defaultextension='.csv')
    df.to_csv (export_file_path, index = None, header=True)

saveAsButton_CSV = tk.Button(text='Export CSV', command=exportCSV, bg='green', fg='white', font=('helvetica', 12, 'bold'))
canvas1.create_window(150, 150, window=saveAsButton_CSV)

root.mainloop()
```
